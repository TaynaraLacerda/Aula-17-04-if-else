IF E ELSE

```python
# if: Define uma condição inicial. Se for verdadeira, o bloco de
#código dentro dele será executado.
#elif (abreviação de else if): Define condições adicionais caso a
#primeira condição (if) seja falsa.
#else: Define um bloco de código que será executado caso todas
#as condições anteriores sejam falsas.
```

```python
# Modelos de IF

idade = 20

if idade >18:
    print('Você é maior de idade! ')


#IF e Else

idade = int(input("Digite sua idade:  "))

if idade >= 18:
  print("Você é maior de idade")
else:
  print("Você é menor de idade")
```

```python
#IF, ELIF, ELSE

idade = 14

if idade <16:
    print("Você é menor de idade")

elif idade >16 and idade <=17:
    print("Voto opcional")

elif idade >=18:
    print("Voto obrigatório")

else:
    print("Nenhuma das opções")
```

```python
# CALCULO COM DESCONTO

vendas = 2100
meta = 1300
meta2 = 2000

if vendas >= meta2:
    bonus= 0.13 * vendas
elif vendas >= meta:
    bonus= 0.1 * vendas
else:
    bonus= 0

print("Bonus:", bonus)
```

```python
# PROJETO LOGIN

usuario = str(input("Digite seu email: "))
senha = str(input("Digite sua senha: "))

if usuario == "felipe123" and senha == "1234":
    print("Acesso permitido")
else:
    print("Usúario Negado!")

usuario = "admin"
senha = "1234"

if usuario == "admin" and senha == "1234":
    print("Acesso permitido!")
else:
    print("Usuário ou senha incorretos.")
```
